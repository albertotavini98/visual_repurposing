```python
#this code iterates on the same image to create a variation of itself
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import os 
import time 
from itertools import product
import random
import io
from util import blender, pianoforte
```

```python
def save_painting(plt, folder):
    now = int(time.time())
    os.makedirs('results\\'+folder, exist_ok=True)
    out_path = 'results\\'+folder+'\\'+str(now)+'.png'
    print('saved ',out_path)
    plt.savefig(out_path, format='png', dpi = 200, bbox_inches='tight', pad_inches=0)
```

```python
width= 1920
dpi = 100  # You can modify the dpi to adjust the quality/resolution of the output image
# Create a figure and a subplot
fig, ax = plt.subplots(figsize=(width/dpi, width/dpi), dpi=dpi)
#fig.patch.set_facecolor('white')
ax.axis('off')



rad1 =  32/width
rad2 = 16/width
rad3 = 8/width
rad4 = 48/width
for i in range (int(1/rad1)):
    for j in range(int(1/rad1)):
        circle = patches.Circle((rad1/2 + i*rad1, rad1/2+ j*rad1), rad1 , edgecolor='tab:blue', facecolor='none')
        ax.add_patch(circle)
        circle = patches.Circle((rad1/2 + i*rad1, rad1/2+ j*rad1), rad2 , edgecolor='tab:red', facecolor='none')
        ax.add_patch(circle)
        circle = patches.Circle((rad1/2 + i*rad1, rad1/2+ j*rad1), rad4 , edgecolor='tab:orange', facecolor='none')
        ax.add_patch(circle)
        """if (i < 4/rad1 and j < 4/rad1 ) or (i >= 4/3*rad1 and j > 4/3*rad1): """

        circle = patches.Circle((rad1/2 + i*rad1, rad1/2+ j*rad1), rad3 , edgecolor='tab:green', facecolor='none')
        ax.add_patch(circle)

save_painting(plt, 'painter')
plt.show()

#we put the content of the figure in a PIL image for future manipulation
buf = io.BytesIO()
plt.savefig(buf, format='png', dpi=200)
buf.seek(0)
img = Image.open(buf)
```

```python
width= 1280
height = 1920
dpi = 100  # You can modify the dpi to adjust the quality/resolution of the output image
# Create a figure and a subplot
fig, ax = plt.subplots(figsize=(width/dpi, height/dpi), dpi=dpi)
#fig.patch.set_facecolor('tab:olive')
ax.set_facecolor('tab:olive')
#ax.axis('off')
colors = ['tab:blue', 'tab:green', 'tab:purple', 'tab:red', 'tab:orange', 'tab:brown','tab:pink', 'tab:cyan']


for i in range(200):
    #we pick a random color
    color = colors[ random.randint(0, len(colors)-1) ]
    #and random coordinates and sizes for each rectangle
    side1 = random.randint(width//8, width//4)
    side2= random.randint(height//12, height//5)
    x = random.randint(0, width )
    y = random.randint(0, height)
    rectangle = patches.Rectangle((x, y), side1, side2, edgecolor=color, facecolor=color)
    ax.add_patch(rectangle)




plt.xlim(0, width)
plt.ylim(0, width)
save_painting(plt, 'painter')
#if i show it before saving it gets emptied (?)
plt.show()

```
