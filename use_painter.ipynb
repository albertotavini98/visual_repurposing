```python
#this code iterates on the same image to create a variation of itself
from PIL import Image
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
import os 
import time 
from itertools import product
import random
import io
from util import blender, pianoforte, painter
```

```python
width = 720
dpi = 100  # You can modify the dpi to adjust the quality/resolution of the output image
# Create a figure and a subplot
fig, ax = plt.subplots(figsize=(width/dpi, width/dpi), dpi=dpi)
#fig.patch.set_facecolor('white')
ax.axis('off')
colors = ['tab:blue', 'tab:green', 'tab:purple', 'tab:red', 'tab:orange', 'tab:brown','tab:pink', 'tab:cyan']


rad = 10

coeff = [random.randint(2, 7) for _ in range(6)]
painter.fillWithCircles(fig, ax,0, width, coeff[0]*rad, coeff[1]*rad, 'tab:pink')
painter.fillWithCircles(fig, ax,0, width, coeff[2]*rad, coeff[3]*rad/2, 'tab:blue')
#painter.fillWithCircles(fig, ax,0, width, coeff[4]*rad, coeff[5]*rad/2, 'tab:cyan')

params = [random.randint(3, 10)*7 for _ in range(12)]
painter.fillWithEllipses(fig, ax, params[0], width, params[1], params[2], params[9] , 'tab:red')
painter.fillWithEllipses(fig, ax, params[3], width, params[4], params[5], params[10] , 'tab:green')
painter.fillWithEllipses(fig, ax, params[6], width, params[7], params[8], params[11] , 'tab:brown')

plt.xlim(0, width)
plt.ylim(0, width)
plt.show()

#we put the content of the figure in a PIL image for future manipulation
buf = io.BytesIO()
plt.savefig(buf, format='jpeg', dpi=200)
buf.seek(0)
img = Image.open(buf)
plt.figure(figsize=(6, 6))
plt.show(img)
```
